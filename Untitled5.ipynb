```python
#Preprocessing
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

#Postprocessing
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
from xgboost import plot_importance
```

```python
X_train = train_df.drop(['SK_ID_CURR', 'TARGET'], axis=1)
y_train = train_df['TARGET']

# Impute missing values with -1000
X_train.fillna(-1000, inplace=True)

# Perform one-hot encoding on categorical features
X_train = pd.get_dummies(X_train)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2)

# Initialize and train the logistic regression model
clf = LogisticRegression()
clf.fit(x_train, y_train)

# accuracy
print("Logistic Regr. Score = ", clf.score(x_test, y_test))
```

```python
clf3 = XGBClassifier()
clf3.fit(x_train, y_train)
print("XGBoost Score = ", clf3.score(x_test, y_test))
```

```python
clf4 = KNeighborsClassifier()
clf4.fit(x_train, y_train)
print("KNN Score = ", clf4.score(x_test, y_test))
```

```python
clf5 = RandomForestClassifier()
clf5.fit(x_train, y_train)
print("Random Forest Score = ", clf5.score(x_test, y_test))
```

```python
# threshold
selection = SelectFromModel(clf3, threshold=0.001, prefit=True)
select_X_train = selection.transform(x_train)
# train model
selection_model = XGBClassifier()
selection_model.fit(select_X_train, y_train)
# eval model
# for X_test
X_test = test_df.drop(['SK_ID_CURR'], axis=1)
X_test.fillna(-1000, inplace=True)          # Fill missing values
X_test = pd.get_dummies(X_test)            # One-hot encode categorical features

# Align columns between X_test and x_train (training data)
# Get missing columns in X_test
missing_cols = set(x_train.columns) - set(X_test.columns)
# Add a missing column in X_test with default value equal to 0
for col in missing_cols:
    X_test[col] = 0
# Ensure the order of column in X_test is in the same order than in x_train
X_test = X_test[x_train.columns]

select_X_test = selection.transform(X_test)
y_pred = selection_model.predict(select_X_test)
```

```python
# Predict probabilities using the model
y_pred = selection_model.predict_proba(select_X_test)
y_pred = pd.DataFrame(y_pred)

# Combine predictions with identifiers for analysis
results = pd.DataFrame()
results['SK_ID_CURR'] = test_df['SK_ID_CURR']  # Unique identifier column from the test dataset
results['Predicted_Probability'] = y_pred.iloc[:, 1]  # Probability of the positive class (fraud)


results_sorted = results.sort_values(by='Predicted_Probability', ascending=False)
print("Top 10 Records with the Highest Predicted Probabilities:")
print(results_sorted.head(10))
print("Last 10 Records with the Highest Predicted Probabilities:")
print(results_sorted.tail(10))


# Visualization: Plot the distribution of predicted probabilities
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(results['Predicted_Probability'], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution of Predicted Probabilities")
plt.xlabel("Predicted Probability")
plt.ylabel("Frequency")
plt.show()

# Visualization: Highlight predictions in a scatter plot (optional)
plt.figure(figsize=(8, 5))
plt.scatter(results['SK_ID_CURR'], results['Predicted_Probability'], alpha=0.6, color='orange')
plt.title("Predicted Probabilities for Test Records")
plt.xlabel("SK_ID_CURR (Unique ID)")
plt.ylabel("Predicted Probability")
plt.show()
```
